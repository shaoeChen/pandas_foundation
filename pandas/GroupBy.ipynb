## Methods on the Groupby Object and DataFrame Columns

## Grouping by Multiple Columns

```python
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
#  可以透過list來做多索引群組
sectors = fortune.groupby(['Sector', 'Industry'])
```

```python
#  可以看的出每個母群組內的子群組數
sectors.size()
```

```python
#  如果ssrs般的多群組呈現出現了
#  sum,min,max,mean...都可以
sectors.sum()
```

```python
#  一樣可以針對單一欄位做加總呈現
sectors['Profits'].sum()
```

## The .agg() Method

```python
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
#  利用.agg達成對不同欄位做不同的統計需求
#  對收入做sum、對利潤做平均，對員工做最大
sectors.agg({
    'Revenue':'sum',
    'Profits':'mean',
    'Employees':'max'
})
```

```python
#  如果單純的提供list，也可以一次性的達成所有欄位的共同統計
sectors.agg(['size','sum','mean'])
```

```python
#  當然結合使用是可以的
sectors.agg({
    'Revenue':['sum','mean'],
    'Profits':'mean',
    'Employees':'max'
})
```

## Iterating through Groups

```python
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
#  如果我們想將21個行業別中收入最高的公司另存一個表的話
#  可以自己建置一個dataframe
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
#  利用迴圈以及.nlargest取得收入最高的那筆資料
for sector, data in sectors:
    highest_revenue = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue)

df
```

```python
#  相同的作法也可以用來取得某個程式中收入最高的那間公司
cities = fortune.groupby('Location')
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue)
df
```
