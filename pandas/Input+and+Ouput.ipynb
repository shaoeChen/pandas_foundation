## Export CSV File with the .to_csv() Method

```python
url = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv'
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
#  將dataframe導出為csv
#  執行之後會匯出在jupyter notebook執行資料夾下
baby_names.to_csv('NYC Baby Name.csv')
```

```python
#  如果不希望csv檔有索引，就可以利用參數來控制
baby_names.to_csv('NYC Baby Name2.csv', index=False)
```

```python
#  如果只需要其中幾個欄位，可利用column來設置
baby_names.to_csv('NYC Baby Name3.csv', index=False, columns=['Year of Birth', 'Count'])
```

```python
#  如果遇到了編碼問題，就利用encoding來設置
baby_names.to_csv('NYC Baby Name4.csv', index=False, columns=['Year of Birth', 'Count'], encoding='utf-8')
```

## Import Excel File Into pandas

```python
#  單一sheet跟導入csv一樣的作法
pd.read_excel('Data - Single Worksheet.xlsx')
```

```python
#  多sheet的時候要設置sheet
#  預設會取第一個sheet
df = pd.read_excel('Data - Multiple Worksheets.xlsx')
df.head()
```

```python
#  透過指定sheet索引或是名稱，可以選擇匯入的資料
df = pd.read_excel('Data - Multiple Worksheets.xlsx', sheet_name=1)
df.head()
```

```python
df = pd.read_excel('Data - Multiple Worksheets.xlsx', sheet_name='Data 2')
df.head()
```

```python
#  也可以一次取回多個sheet的資料，格式就會變成dict
df = pd.read_excel('Data - Multiple Worksheets.xlsx', sheet_name=[0, 1])
df
```

```python
len(df)
#  df[1]
```

```python
#  索引值會依你匯入的時候使用索引或是sheet名稱而變動
df = pd.read_excel('Data - Multiple Worksheets.xlsx', sheet_name=['Data 1', 'Data 2'])
df
```

```python
#  如果參數給none, 預設會全數匯入
df = pd.read_excel('Data - Multiple Worksheets.xlsx', sheet_name=None)
df
```

## Export Excel File

```python
url = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv'
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
girls = baby_names[baby_names['Gender']=='FEMALE']
boys = baby_names[baby_names['Gender']=='MALE']
```

```python
#  這時候還沒有真正的產生檔案，只是new一個物件
excel_file = pd.ExcelWriter('BabyNames.xlsx')
```

```python
#  參數columns可以選擇要輸出的欄位
#  這時候檔案還沒有真正的被產生，只是將資料塞進去上面的物件
girls.to_excel(excel_file, sheet_name='Girls', index=False)
boys.to_excel(excel_file, sheet_name='Boys', index=False, columns=['Count', 'Rank'])
```

```python
#  這時候真正的產生了
excel_file.save()
```
